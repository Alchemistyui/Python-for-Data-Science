```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./world-development-indicators/Indicators.csv')
data.head()
```

```python
data.shape
data.head(10)
```

```python
len(data.CountryName.unique().tolist())
```

```python
hist_indicator = 'CO2 emissions \(metric'
hist_country = 'USA'

indicator_mask = data['IndicatorName'].str.contains(hist_indicator)
country_mask = data['CountryCode'].str.contains(hist_country)

usa_co2 = data[indicator_mask & country_mask]
len(usa_co2)
usa_co2.head()
```

```python
years = usa_co2.Year.values

co2 = usa_co2.Value.values

plt.bar(years, co2)
plt.show()
```

```python
usa_co2.IndicatorName.iloc[0]
```

```python
plt.plot(years, co2)

plt.xlabel('Year')
plt.ylabel(usa_co2.IndicatorName.iloc[0])

plt.title('co2 emmision in usa')

plt.axis([1959,2011, 0,25])

plt.show()
```

```python
hist_data = usa_co2.Value.values
len(hist_data)
```

```python
plt.hist(hist_data, 10, density=False, facecolor='green')

plt.xlabel(usa_co2.IndicatorName.iloc[0])
plt.ylabel('# of years')

plt.title('histogram example')

plt.grid(True)

plt.show()
```

```python
hist_indicator = 'CO2 emissions \(metric'
hist_year = 2011

indicator_mask = data['IndicatorName'].str.contains(hist_indicator)
year_mask = data['Year'].isin([hist_year])

co2_2011 = data[indicator_mask & year_mask]
co2_2011
```

```python
data['Year'].dtypes
```

```python
fig, axis = plt.subplots()

axis.annotate('USA', 
             xy=(18,5), xycoords='data',
             xytext=(18,30), textcoords='data',
             arrowprops=dict(arrowstyle='->',
                            connectionstyle='arc3'))
plt.hist(co2_2011.Value, 10, density=False, facecolor='green')

plt.xlabel(co2_2011.IndicatorName.iloc[0])
plt.ylabel('# of countries')

plt.title('histogram for co2 countries 2011')

plt.grid(True)

plt.show()
```

```python
hist_indicator = 'GDP per capita \(constant 2005'
hist_country = 'USA'

indicator_mask = data['IndicatorName'].str.contains(hist_indicator)
country_mask = data['CountryCode'].str.contains(hist_country)

usa_gdp = data[indicator_mask & country_mask]
```
